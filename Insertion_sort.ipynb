```python
import timeit
import numpy as np
import matplotlib.pyplot as plt

def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
    # Define input sizes
sizes = [5, 10, 20, 50, 100, 500, 1000, 5000]
times = []

for size in sizes:
    arr = np.random.randint(1, 10000, size)  # Generate a random array
    time_taken = timeit.timeit(lambda: insertion_sort(arr.copy()), number=5) / 5  # Average over 5 runs
    times.append(time_taken)

#Plot Execution Time vs Input Size
plt.plot(sizes, times, marker='o', linestyle='-', linewidth=2, markersize=6)
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Insertion Sort Performance")
plt.grid(True)
plt.show()
import platform
import psutil

#Benchmarking
print(f"Processor: {platform.processor()}")
print(f"CPU Cores: {psutil.cpu_count(logical=False)}")
print(f"RAM: {round(psutil.virtual_memory().total / (1024 * 1024 * 1024), 2)} GB")

```
